### 네이버 영화 평점: 2019년 9월 27일

```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
# https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=20190927
url_base = "https://movie.naver.com"
url_syb = "/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=20190927"

page = urlopen(url_base + url_syb)
soup = BeautifulSoup(page, 'html.parser') 
print(soup)
```

```python
movie_name = [soup.find_all('div', 'tit5')[n].a.string for n in range(0,36)]
movie_name
```

```python
movie_point = [soup.find_all('td', 'point')[n].string for n in range(0,36)]
movie_point
```

```python
date = pd.date_range('2019-6-1', freq='D', periods=90)
date
```

```python
# date에 해당하는 기간에 모든 영화 이름, 평점, 개봉일을 불러온다.
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = "http://movie.naver.com/" + "movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date': movie_date,
                     'name': movie_name,
                     'point': movie_point})
movie.head()
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
```

```python
movie.info()
```

```python
import numpy as np
movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
tmp = movie.query('name == ["알라딘"]')
tmp 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(tmp, loc='best')
plt.grid()
plt.show()
```

```python
# https://music.naver.com/listen/history/index.nhn?type=DOMESTIC_V2&year=2019&month=09&week=3
import requests
from bs4 import BeautifulSoup

url = 'https://music.naver.com/listen/history/index.nhn?type=DOMESTIC_V2&year=2019&month=09&week=3'
html_music = requests.get(url).text
soup_music = BeautifulSoup(html_music, 'html.parser')
titles = soup_music.select('a.title span.ellipsis')

music_title = [title.get_text() for title in titles]
len(music_title)
```

```python
artists = soup_music.select('td._artist a')
music_artist = [artist.get_text().strip() for artist in artists]
len(music_artist)
```
